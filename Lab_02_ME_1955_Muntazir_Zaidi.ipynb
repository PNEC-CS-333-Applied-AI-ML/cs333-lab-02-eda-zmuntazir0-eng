- Which method did you use (IQR or Modified Z-score)?
- Why?

I used the IQR meethod instead of z-score because IQR is a better option for negatively skewed data whereas the Z-score, which assumes a normal distribution and can be heavily influenced by outliers themselves.


- How many outliers were removed?

47 outliers (180 - 133) were removed from the dataset.


- Which parameters contain significant outliers?

vib_table had the most outliers


- How did it affect statistics?

removing outliers made my data less skewed as can be seen from the variance i have found of df2 at the end
